```python
import re
import pandas as pd

# SQL 파일 읽기
with open("Data/index.sql", "r", encoding="utf-8") as file:
    sql_content = file.read()

# INSERT INTO 구문에서 VALUES 부분 추출
matches = re.findall(r"INSERT INTO `[^`]+` \([^)]+\) VALUES\s*(.+);", sql_content, re.DOTALL)

# 데이터 정리
data_rows = []
for match in matches:
    # 여러 행 분리
    rows = re.findall(r"\((.*?)\)", match)
    for row in rows:
        # 쉼표로 분리하되, 작은따옴표와 괄호 안의 쉼표는 제외
        split_row = re.split(r",(?=(?:[^']*'[^']*')*[^']*$)", row)
        # 작은따옴표와 양쪽 공백 제거
        cleaned_row = [col.strip().strip("'") for col in split_row]
        # 유효한 데이터만 추가
        if len(cleaned_row) == 5:  # 정확히 5개의 열이 있는 데이터만 추가
            data_rows.append(cleaned_row)

# 열 이름 추가
columns = ['rec_id', 'url', 'website', 'result', 'created_date']

# DataFrame 생성
df = pd.DataFrame(data_rows, columns=columns)
df = df.drop_duplicates(subset=df.columns.tolist(), keep=False)

# CSV 파일로 저장 (컬럼 포함)
csv_file_path = "Data/mendely_url.csv"
df.to_csv(csv_file_path, index=False, encoding="utf-8")

print(f"CSV 파일이 성공적으로 저장되었습니다: {csv_file_path}")
```
